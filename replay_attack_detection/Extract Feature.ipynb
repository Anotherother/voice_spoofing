```python
import numpy as np
import scipy
from scipy import stats

import pandas as pd
from scipy.io import wavfile

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

import librosa
from librosa import display

%matplotlib inline
```

```python
def snr(data):  
    singleChannel = data
    try:
        singleChannel = numpy.sum(data, axis=1)
    except:
        pass
    norm = singleChannel / float(max(np.amax(singleChannel), -1 * np.amin(singleChannel)))

    return stats.signaltonoise(norm)

def ex(X):
    mu = np.sum((X-np.mean(X))**4)/(X.shape[0]-1)
    return mu/(sigmaAudio(X)**4) -3

def sigmaAudio(X):
    return np.sqrt(sum((X - np.mean(X))**2)/len(X)) #sigma 

def rangeSignal(signal):
    return np.max(signal) - np.min(signal)

def computeMFCC(audio, sr):
    mcep = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=20).T
    return mcep
```

```python
def comuteParams(setName, typeAudio):
    dataframe = pd.read_csv(str(setName) + '_label.csv')
    
    sigmaFeature = []
    signalToNoiseRatio = []
    excess = []
    rangeSample = []
    if (typeAudio == 'spoof'):
        nameFile = dataframe[dataframe.Type=='spoof'].FileID
    elif (typeAudio == 'original'):
        nameFile = dataframe[dataframe.Type=='genuine'].FileID
    else:
        print 'Error'
        
    for i in nameFile:
        filename = '../asvspoof2017/ASVspoof2017_train_dev/wav/' + str(setName)+'/' + i
        audio, sr = librosa.load(filename)

        
        sigmaFeature.append(sigmaAudio(audio))
        signalToNoiseRatio.append(snr(audio))
        excess.append(ex(audio))
        rangeSample.append(rangeSignal(audio))

        
    return sigmaFeature, signalToNoiseRatio, excess, rangeSample
```

```python
sigmaVectorOriginal,snrOriginal, exOriginal, mfccOriginal = comuteParams('train', 'original')
```
